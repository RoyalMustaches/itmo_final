```python
import pandas as pd
import numpy as np

import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# 1. Подготовка данных
# =========================

for col in ['gmv_with_markup']:
  
    s = pd.to_numeric(df[col], errors='coerce')
    valid = s.dropna()

    if valid.empty:
        print(f"Колонка {col} не содержит числовых значений, шаг пропущен.")
        continue

    q1 = valid.quantile(0.25)
    q3 = valid.quantile(0.75)
    iqr = q3 - q1

    if pd.isna(iqr) or iqr == 0:
        print(f"Для {col} IQR = 0/NaN, выбросы не удалялись. Строк: {len(df)}")
        continue

    upper_whisker = q3 + 1.5 * iqr
    keep_mask = s.isna() | (s <= upper_whisker)

    before = len(df)
    df = df[keep_mask].copy()
    print(f"После очистки {col}: {len(df)} строк (удалено {before - len(df)})")

df_model = df.copy()

df_model.columns = (
    df_model.columns
    .str.replace('\ufeff', '', regex=False)
    .str.strip()
)

df_model = df_model[df_model['gmv_with_markup'] > 0].copy()

df_model = df_model.dropna(
    subset=[
        'gmv_with_markup',
        'markup',
        'cb_percent'
    ]
).copy()

df_model = df_model[df_model['markup'].isin([0, 1, 5, 9])].copy()
df_model['markup'] = df_model['markup'].astype(int)

# =========================
# 2. Флаги наценки
# =========================
# Базовая группа: markup = 0
# Сами флаги отдельно в модель не включаем

df_model['markup_1'] = (df_model['markup'] == 1).astype(int)
df_model['markup_5'] = (df_model['markup'] == 5).astype(int)
df_model['markup_9'] = (df_model['markup'] == 9).astype(int)

# =========================
# 3. cb_percent, квадрат и взаимодействия
# =========================

df_model['cb_percent_sq'] = df_model['cb_percent'] ** 2

df_model['cb_x_markup_1'] = df_model['cb_percent'] * df_model['markup_1']
df_model['cb_x_markup_5'] = df_model['cb_percent'] * df_model['markup_5']
df_model['cb_x_markup_9'] = df_model['cb_percent'] * df_model['markup_9']

# =========================
# 4. Линейная регрессия
# =========================

formula = """
gmv_with_markup ~ cb_percent
                + cb_percent_sq
                + cb_x_markup_1
                + cb_x_markup_5
                + cb_x_markup_9
"""

if 'client_id' in df_model.columns:
    model_gmv_cb_sq = smf.ols(
        formula=formula,
        data=df_model
    ).fit(
        cov_type='cluster',
        cov_kwds={'groups': df_model['client_id']}
    )
else:
    model_gmv_cb_sq = smf.ols(
        formula=formula,
        data=df_model
    ).fit(cov_type='HC3')

print(model_gmv_cb_sq.summary())

# =========================
# 5. Таблица коэффициентов
# =========================

coef_table = model_gmv_cb_sq.summary2().tables[1].reset_index()
coef_table = coef_table.rename(columns={
    'index': 'term',
    'Coef.': 'coef',
    'Std.Err.': 'std_err',
    '[0.025': 'ci_low',
    '0.975]': 'ci_high',
    'P>|z|': 'p_value'
})

display(coef_table)

# =========================
# 6. Совместные тесты
# =========================

print('Совместный тест взаимодействий cb_percent × markup:')
print(
    model_gmv_cb_sq.wald_test(
        'cb_x_markup_1 = 0, cb_x_markup_5 = 0, cb_x_markup_9 = 0'
    )
)

print('Тест квадратичного эффекта cb_percent:')
print(
    model_gmv_cb_sq.wald_test(
        'cb_percent_sq = 0'
    )
)

# =========================
# 7. Матрица корреляций предикторов
# =========================

predictors = [
    'cb_percent',
    'cb_percent_sq',
    'cb_x_markup_1',
    'cb_x_markup_5',
    'cb_x_markup_9'
]

X = df_model[predictors].dropna().copy()

corr_matrix = X.corr()

display(corr_matrix)

plt.figure(figsize=(9, 7))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=0.7,
    annot_kws={"size": 10}
)

plt.title('Матрица корреляций предикторов модели')
plt.tight_layout()
plt.show()

# =========================
# 8. VIF
# =========================

X_vif = sm.add_constant(X)

vif_table = pd.DataFrame({
    'variable': X_vif.columns,
    'VIF': [
        variance_inflation_factor(X_vif.values, i)
        for i in range(X_vif.shape[1])
    ]
})

vif_table = vif_table[vif_table['variable'] != 'const'].copy()
vif_table = vif_table.sort_values('VIF', ascending=False)

display(vif_table)

# =========================
# 9. Предсказания модели
# =========================

cb_min = df_model['cb_percent'].quantile(0.01)
cb_max = df_model['cb_percent'].quantile(0.99)

cb_grid = np.linspace(cb_min, cb_max, 100)

pred_grid = pd.DataFrame([
    {
        'cb_percent': cb,
        'markup': markup
    }
    for cb in cb_grid
    for markup in [0, 1, 5, 9]
])

pred_grid['cb_percent_sq'] = pred_grid['cb_percent'] ** 2

pred_grid['markup_1'] = (pred_grid['markup'] == 1).astype(int)
pred_grid['markup_5'] = (pred_grid['markup'] == 5).astype(int)
pred_grid['markup_9'] = (pred_grid['markup'] == 9).astype(int)

pred_grid['cb_x_markup_1'] = pred_grid['cb_percent'] * pred_grid['markup_1']
pred_grid['cb_x_markup_5'] = pred_grid['cb_percent'] * pred_grid['markup_5']
pred_grid['cb_x_markup_9'] = pred_grid['cb_percent'] * pred_grid['markup_9']

pred = model_gmv_cb_sq.get_prediction(pred_grid).summary_frame(alpha=0.05)

pred_grid['predicted_gmv'] = pred['mean']
pred_grid['ci_low'] = pred['mean_ci_lower']
pred_grid['ci_high'] = pred['mean_ci_upper']

display(pred_grid)

# =========================
# 10. График предсказаний модели
# =========================

plt.figure(figsize=(12, 8))

pred_palette = {
    0: '#1b9e77',
    1: '#d95f02',
    5: '#7570b3',
    9: '#e7298a'
}

sns.lineplot(
    data=pred_grid,
    x='cb_percent',
    y='predicted_gmv',
    hue='markup',
    hue_order=[0, 1, 5, 9],
    palette=pred_palette,
    linewidth=2.8
)

for markup_value in [0, 1, 5, 9]:
    temp = pred_grid[pred_grid['markup'] == markup_value]

    plt.fill_between(
        temp['cb_percent'],
        temp['ci_low'],
        temp['ci_high'],
        color=pred_palette[markup_value],
        alpha=0.20
    )

plt.title('Предсказанный GMV по кэшбэку, квадрату кэшбэка и наценке')
plt.xlabel('Кэшбэк, %')
plt.ylabel('Предсказанный gmv_with_markup')
plt.legend(title='Наценка, %')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# =========================
# 11. Описательный bootstrap-график средних GMV
# =========================

plot_data = df_model[
    [
        'markup',
        'cb_percent',
        'gmv_with_markup'
    ]
].dropna().copy()

cb_markup_avg = (
    plot_data
    .groupby(['markup', 'cb_percent'], as_index=False)
    .agg(
        gmv_mean=('gmv_with_markup', 'mean'),
        n_orders=('gmv_with_markup', 'size')
    )
)

display(cb_markup_avg)

def bootstrap_poly_fit(x, y, x_fit, degree=2, n_bootstraps=1000, confidence=0.95):
    x = np.asarray(x)
    y = np.asarray(y)

    n = len(x)
    bootstrap_fits = np.zeros((n_bootstraps, len(x_fit)))

    for i in range(n_bootstraps):
        indices = np.random.choice(n, n, replace=True)

        x_resample = x[indices]
        y_resample = y[indices]

        if len(np.unique(x_resample)) <= degree:
            bootstrap_fits[i, :] = np.nan
            continue

        coefficients_resample = np.polyfit(x_resample, y_resample, degree)
        polynomial_resample = np.poly1d(coefficients_resample)

        bootstrap_fits[i, :] = polynomial_resample(x_fit)

    lower_percentile = (1 - confidence) / 2 * 100
    upper_percentile = (1 + confidence) / 2 * 100

    lower_bound = np.nanpercentile(bootstrap_fits, lower_percentile, axis=0)
    upper_bound = np.nanpercentile(bootstrap_fits, upper_percentile, axis=0)

    return lower_bound, upper_bound

plt.figure(figsize=(12, 8))

palette = {
    0: 'purple',
    1: 'blue',
    5: 'green',
    9: 'red'
}

degree = 2

for markup_value in [0, 1, 5, 9]:
    temp = cb_markup_avg[cb_markup_avg['markup'] == markup_value].copy()

    if temp['cb_percent'].nunique() <= degree:
        print(f"Для markup={markup_value}% слишком мало уникальных значений cb_percent")
        continue

    x_data = temp['cb_percent'].values
    y_data = temp['gmv_mean'].values

    plt.scatter(
        x_data,
        y_data,
        color=palette[markup_value],
        alpha=0.85,
        edgecolor="white",
        linewidth=0.5,
        label=f'Наценка {markup_value}%: средний GMV'
    )

    coefficients = np.polyfit(x_data, y_data, degree)
    polynomial = np.poly1d(coefficients)

    x_fit = np.linspace(x_data.min(), x_data.max(), 100)
    y_fit = polynomial(x_fit)

    plt.plot(
        x_fit,
        y_fit,
        color=palette[markup_value],
        linewidth=2.8
    )

    lower_bound, upper_bound = bootstrap_poly_fit(
        x=x_data,
        y=y_data,
        x_fit=x_fit,
        degree=degree,
        n_bootstraps=1000,
        confidence=0.95
    )

    plt.fill_between(
        x_fit,
        lower_bound,
        upper_bound,
        color=palette[markup_value],
        alpha=0.15
    )

plt.title("Средний GMV по кэшбэку и уровню наценки")
plt.xlabel("Кэшбэк, %")
plt.ylabel("Средний gmv_with_markup")
plt.legend(title='Наценка')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Автоподготовка, если ячейка запускается отдельно
if 'df_model' not in globals():
    if 'df' not in globals():
        raise NameError("Не найден ни df_model, ни df. Сначала выполните первую ячейку с загрузкой данных.")

    df_model = df.copy()
    df_model.columns = (
        df_model.columns
        .str.replace('\ufeff', '', regex=False)
        .str.strip()
    )
    df_model = df_model[df_model['gmv_with_markup'] > 0].copy()
    df_model = df_model.dropna(subset=['gmv_with_markup', 'markup', 'cb_percent']).copy()
    df_model = df_model[df_model['markup'].isin([0, 1, 5, 9])].copy()
    df_model['markup'] = df_model['markup'].astype(int)

if 'markup_1' not in df_model.columns:
    df_model['markup_1'] = (df_model['markup'] == 1).astype(int)
if 'markup_5' not in df_model.columns:
    df_model['markup_5'] = (df_model['markup'] == 5).astype(int)
if 'markup_9' not in df_model.columns:
    df_model['markup_9'] = (df_model['markup'] == 9).astype(int)
if 'cb_percent_sq' not in df_model.columns:
    df_model['cb_percent_sq'] = df_model['cb_percent'] ** 2
if 'cb_x_markup_1' not in df_model.columns:
    df_model['cb_x_markup_1'] = df_model['cb_percent'] * df_model['markup_1']
if 'cb_x_markup_5' not in df_model.columns:
    df_model['cb_x_markup_5'] = df_model['cb_percent'] * df_model['markup_5']
if 'cb_x_markup_9' not in df_model.columns:
    df_model['cb_x_markup_9'] = df_model['cb_percent'] * df_model['markup_9']

import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# 1. Подготовка переменных
# =========================

df_corr = df_model.copy()

df_corr.columns = (
    df_corr.columns
    .str.replace('\ufeff', '', regex=False)
    .str.strip()
)

# Если переменные еще не созданы, создаем их
if 'cb_percent_sq' not in df_corr.columns:
    df_corr['cb_percent_sq'] = df_corr['cb_percent'] ** 2

if 'cb_x_markup_1' not in df_corr.columns:
    df_corr['cb_x_markup_1'] = df_corr['cb_percent'] * (df_corr['markup'].astype(str) == '1').astype(int)

if 'cb_x_markup_5' not in df_corr.columns:
    df_corr['cb_x_markup_5'] = df_corr['cb_percent'] * (df_corr['markup'].astype(str) == '5').astype(int)

if 'cb_x_markup_9' not in df_corr.columns:
    df_corr['cb_x_markup_9'] = df_corr['cb_percent'] * (df_corr['markup'].astype(str) == '9').astype(int)

columns_to_p_value_corr = [
    'cb_percent',
    'cb_percent_sq',
    'cb_x_markup_1',
    'cb_x_markup_5',
    'cb_x_markup_9'
]

df_corr = df_corr[columns_to_p_value_corr].apply(pd.to_numeric, errors='coerce')

# =========================
# 2. Функция для p-value
# =========================

def pearsonr_pval(x, y):
    valid = pd.concat([pd.Series(x), pd.Series(y)], axis=1).dropna()

    if valid.shape[0] < 3:
        return np.nan

    if valid.iloc[:, 0].nunique() <= 1 or valid.iloc[:, 1].nunique() <= 1:
        return np.nan

    return st.pearsonr(valid.iloc[:, 0], valid.iloc[:, 1])[1]

# =========================
# 3. Матрица p-value
# =========================

pval_matrix = df_corr.corr(method=pearsonr_pval)

# Для диагонали p-value не имеет смысла
pval_matrix = pval_matrix.where(~np.eye(len(pval_matrix), dtype=bool), np.nan)

# Формат подписей в научной записи
annot_matrix = pval_matrix.map(lambda x: f'{x:.1e}' if pd.notna(x) else '')

# =========================
# 4. Heatmap
# =========================

plt.figure(figsize=(9, 6))

sns.heatmap(
    pval_matrix,
    linewidths=0.5,
    cmap='PRGn_r',
    cbar_kws={'shrink': 0.8, 'label': 'p-value'},
    annot=annot_matrix,
    fmt=''
)

plt.title('Матрица p-value Pearson-корреляций между предикторами')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Импорты на случай отдельного запуска ячейки
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm

# =========================
# Проверка устойчивости при добавлении контролей
# =========================

base_formula = """
gmv_with_markup ~ cb_percent
                + cb_percent_sq
                + cb_x_markup_1
                + cb_x_markup_5
                + cb_x_markup_9
"""

candidate_controls = [
    'gender_cd',
    'age_group',
    'income_group',
    'lifestyle___super_cnt',
    'lifestyle___super_sum'
]

available_controls = [
    col for col in candidate_controls
    if col in df_model.columns
]

formulas = {
    'base': base_formula,
    'with_gender_age': base_formula + ' + C(gender_cd) + C(age_group)'
        if {'gender_cd', 'age_group'}.issubset(df_model.columns)
        else base_formula,
    'with_income': base_formula + ' + C(income_group)'
        if 'income_group' in df_model.columns
        else base_formula,
    'with_history': base_formula + ' + lifestyle___super_cnt + lifestyle___super_sum'
        if {'lifestyle___super_cnt', 'lifestyle___super_sum'}.issubset(df_model.columns)
        else base_formula,
    'full_controls': base_formula + ''.join(
        [
            ' + C(gender_cd)' if 'gender_cd' in df_model.columns else '',
            ' + C(age_group)' if 'age_group' in df_model.columns else '',
            ' + C(income_group)' if 'income_group' in df_model.columns else '',
            ' + lifestyle___super_cnt' if 'lifestyle___super_cnt' in df_model.columns else '',
            ' + lifestyle___super_sum' if 'lifestyle___super_sum' in df_model.columns else ''
        ]
    )
}

control_results = []

for model_name, f in formulas.items():
    temp = df_model.copy()

    needed_cols = [
        'gmv_with_markup',
        'cb_percent',
        'cb_percent_sq',
        'cb_x_markup_1',
        'cb_x_markup_5',
        'cb_x_markup_9'
    ]

    temp = temp.dropna(subset=[col for col in needed_cols if col in temp.columns]).copy()

    base_fit = smf.ols(formula=f, data=temp).fit()

    if 'client_id' in temp.columns and temp['client_id'].nunique() > 1:
        used_idx = base_fit.model.data.row_labels
        groups = temp.loc[used_idx, 'client_id']
        model = base_fit.get_robustcov_results(
            cov_type='cluster',
            groups=groups
        )
    else:
        model = base_fit.get_robustcov_results(cov_type='HC3')

    param_names = model.model.exog_names
    params = pd.Series(model.params, index=param_names)
    pvalues = pd.Series(model.pvalues, index=param_names)

    for term in [
        'cb_percent',
        'cb_percent_sq',
        'cb_x_markup_1',
        'cb_x_markup_5',
        'cb_x_markup_9'
    ]:
        if term in params.index:
            control_results.append({
                'model': model_name,
                'n_obs': int(model.nobs),
                'term': term,
                'coef': params[term],
                'p_value': pvalues[term],
                'r_squared': model.rsquared,
                'aic': model.aic
            })

control_results = pd.DataFrame(control_results)

display(control_results)
```
